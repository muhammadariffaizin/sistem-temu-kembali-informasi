```python
import pandas as pd
import numpy as np
```

```python
bungs = pd.read_csv("bungs.csv")
catatanali07 = pd.read_csv("catatanali07.csv")
bungs_labelled = pd.read_csv("bungs_labelled.csv", dtype={'Id':np.int64})
catatanali07_labelled = pd.read_csv("catatanali07_labelled.csv", dtype={'Id':np.int64})
```

```python
bungs_count_labelled = bungs_labelled[['KS', 'KH', 'DH', 'TDH', 'SWA', 'LO', 'AK']].count()
catatanali07_count_labelled = catatanali07_labelled[['KS', 'KH', 'DH', 'TDH', 'SWA', 'LO', 'AK']].count()
display(bungs_count_labelled)
print("Total : ", bungs_count_labelled.sum())
display(catatanali07_count_labelled)
print("Total : ", catatanali07_count_labelled.sum())
```

```python
bungs_labelled_drop = bungs_labelled[['Id', 'KS', 'KH', 'DH', 'TDH', 'SWA', 'LO', 'AK']]
catatanali07_labelled_drop = catatanali07_labelled[['Id', 'KS', 'KH', 'DH', 'TDH', 'SWA', 'LO', 'AK']]

display(bungs.head())
display(catatanali07.head())
display(bungs_labelled_drop.head())
display(catatanali07_labelled_drop.head())
```

```python
bungs_merged = bungs.merge(bungs_labelled_drop, left_on="Id", right_on="Id", how="left")
catatanali07_merged = catatanali07.merge(catatanali07_labelled_drop, left_on="Id", right_on="Id", how="left")
```

```python
bungs_merged.to_csv("bungs_merged_labelled.csv")
catatanali07_merged.to_csv("catatanali07_merged_labelled.csv")
```

```python
bungs_count_merged = bungs_merged[['KS', 'KH', 'DH', 'TDH', 'SWA', 'LO', 'AK']].count()
catatanali07_count_merged = catatanali07_merged[['KS', 'KH', 'DH', 'TDH', 'SWA', 'LO', 'AK']].count()
display(bungs_count_merged)
print("Total : ", bungs_count_merged.sum())
display(catatanali07_count_merged)
print("Total : ", catatanali07_count_merged.sum())
```
