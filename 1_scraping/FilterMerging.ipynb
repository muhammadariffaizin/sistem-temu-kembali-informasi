```python
import pandas as pd
import json
```

```python
import snscrape.modules.twitter as sntwitter

# get user
user_catatanali07 = sntwitter.TwitterUserScraper("catatanali07").entity
user_bungs = sntwitter.TwitterUserScraper("__bungs").entity

user1_data = [user_catatanali07.id, user_catatanali07.username, user_catatanali07.location,
              user_catatanali07.followersCount, user_catatanali07.friendsCount, user_catatanali07.favouritesCount,
              user_catatanali07.statusesCount]
user2_data = [user_bungs.id, user_bungs.username, user_bungs.location,
              user_bungs.followersCount, user_bungs.friendsCount, user_bungs.favouritesCount,
              user_bungs.statusesCount]

display(user1_data)
display(user2_data)
```

# CatatanAli07

## Raw Data

```python
tweet = pd.read_csv("tweet_catatanali07_base.csv")
users = pd.read_csv("users_catatanali07_base.csv")
display(tweet)
display(users)
```

```python
tweet_filtered = tweet.filter(items=["id", "date", "url", "content", "likeCount", "replyCount", "retweetCount", "retweetedTweet"])
tweet_filtered = tweet_filtered.assign(user_id=user1_data[0], user_name=user1_data[1], user_location=user1_data[2], 
                      follower_count=user1_data[3], following_count=user1_data[4], likes_count=user1_data[5], tweets_count=user1_data[6])
tweet_filtered = tweet_filtered.rename(columns={
                                        tweet_filtered.columns[0]: "Id", 
                                        tweet_filtered.columns[1]: "Tweet Time", 
                                        tweet_filtered.columns[2]: "Url", 
                                        tweet_filtered.columns[3]: "Tweet Text", 
                                        tweet_filtered.columns[4]: "Likes Count", 
                                        tweet_filtered.columns[5]: "Reply Count", 
                                        tweet_filtered.columns[6]: "Retweet Count", 
                                        tweet_filtered.columns[7]: "Retweeted Tweet", 
                                        tweet_filtered.columns[8]: "User Id", 
                                        tweet_filtered.columns[9]: "User Name", 
                                        tweet_filtered.columns[10]: "User Location", 
                                        tweet_filtered.columns[11]: "User Follower Count", 
                                        tweet_filtered.columns[12]: "User Following Count", 
                                        tweet_filtered.columns[13]: "User Likes Count", 
                                        tweet_filtered.columns[14]: "User Tweets Count", 
                                    })

display(tweet_filtered)
```

```python
base_data = tweet_filtered
base_data.to_csv('merge_catatanali07_base.csv', index=False)
```

## Replies Data

```python
tweet = pd.read_csv("tweet_catatanali07_replies.csv")
users = pd.read_csv("users_catatanali07_replies.csv")
display(tweet)
display(users)
```

```python
tweet_filtered = tweet.filter(items=["id", "date", "url", "content", "likeCount", "replyCount", "retweetCount", "retweetedTweet"])

username_series = tweet.user.str.split("', '")
username_series = username_series.apply(lambda x: x[0].split("': '")[1])
tweet_filtered = tweet_filtered.assign(username_user=username_series)

tweet_filtered = tweet_filtered.merge(users, left_on="username_user", right_on="Username", how="left")
# tweet_filtered = tweet_filtered.assign(user_id=user1_data[0], user_name=user1_data[1], user_location=user1_data[2], 
#                       follower_count=user1_data[3], following_count=user1_data[4], likes_count=user1_data[5], tweets_count=user1_data[6])
tweet_filtered = tweet_filtered.drop('username_user', axis=1)
tweet_filtered = tweet_filtered.rename(columns={
                                        tweet_filtered.columns[0]: "Id", 
                                        tweet_filtered.columns[1]: "Tweet Time", 
                                        tweet_filtered.columns[2]: "Url", 
                                        tweet_filtered.columns[3]: "Tweet Text", 
                                        tweet_filtered.columns[4]: "Likes Count", 
                                        tweet_filtered.columns[5]: "Reply Count", 
                                        tweet_filtered.columns[6]: "Retweet Count", 
                                        tweet_filtered.columns[7]: "Retweeted Tweet", 
                                        tweet_filtered.columns[8]: "User Id", 
                                        tweet_filtered.columns[9]: "User Name", 
                                        tweet_filtered.columns[10]: "User Follower Count", 
                                        tweet_filtered.columns[11]: "User Following Count", 
                                        tweet_filtered.columns[12]: "User Tweets Count", 
                                        tweet_filtered.columns[13]: "User Likes Count", 
                                        tweet_filtered.columns[14]: "User Location", 
                                    })

display(tweet_filtered)
```

```python
replies_data = tweet_filtered
replies_data.to_csv('merge_catatanali07_replies.csv', index=False)
```
